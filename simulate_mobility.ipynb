```python
import numpy as np
import pandas as pd
import pickle
```

```python
cluster_id = 0; time_interval = 24; week_or_weekend = 'weekend'
cluster_user = np.load(week_or_weekend+'_user/cluster_'+str(cluster_id)+'.npy')
transition_matrix = np.load(week_or_weekend+'_user/transition_matrix.npy')
initial_state_list = pd.read_csv(week_or_weekend+'_user/initial_state.csv')
initial_state_list = initial_state_list[initial_state_list['t']==0]
transition_matrix_cluster = transition_matrix[cluster_id,:,:,:]

state_user = {}
# what does cluster_label means? 'average' visiting place type for each cluster (0.5h interval)?
for user_id in cluster_user:
    # initial state marix
    # intitial_vector = np.array([0.6,0.2,0.1,0.1])
    # initial_state = np.random.choice(range(4), 1, p=intitial_vector)[0]
    try:
        initial_label = initial_state_list[initial_state_list['uid']==user_id]['label'].values[0]
    except:
        # print(initial_state_list[initial_state_list['uid']==user_id]['label'].values)
        initial_label = 'home'
    if initial_label == 'home':
        initial_state = 0
    elif initial_label == 'work':
        initial_state = 1
    elif initial_label == 'E':
        initial_state = 2
    elif initial_label == 'R':
        initial_state = 3
    
    state_hour = []
    for hour in range(time_interval):
        # if we ignore different time periods, we can generate the chain.
        if hour>=22 or hour<8:
            transition_vector = transition_matrix_cluster[0,initial_state,:]
        elif hour>=8 and hour<17:
            transition_vector = transition_matrix_cluster[1,initial_state,:]
        elif hour>=17 and hour<22:
            transition_vector = transition_matrix_cluster[2,initial_state,:]
        
        # each row should sum to 1?
        transition_vector = transition_vector/np.sum(transition_vector)
        transit_state = np.random.choice(range(4), 1, p=transition_vector)[0]
        initial_state = int(transit_state)
        state_hour.append(transit_state)
    state_user[user_id] = state_hour
    
    # if we consider three time periods, we need arrival and departure time of each stay type (h, w, or, oe)?
pickle.dump(state_user, open(week_or_weekend+'_user/user_mobility.pkl', 'wb'))
```
